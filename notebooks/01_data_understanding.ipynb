```python
df[target].value_counts().plot(kind="bar")

plt.title("Default vs Non-Default Borrowers")
plt.xlabel("Default Status")
plt.ylabel("Number of Borrowers")
plt.show()
```

```python
## Initial Business Observations

The dataset contains borrower-level financial and behavioural variables that can be used for credit risk modelling.

The target variable is `SeriousDlqin2yrs`, which indicates whether a borrower experienced serious delinquency within two years.

The default rate represents the baseline risk level of the borrower population. A credit risk model should help identify borrowers with higher-than-average and lower-than-average default risk.

The next step is to perform exploratory data analysis to understand which borrower characteristics are most strongly associated with default.
```
